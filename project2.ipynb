### Visualize the distribution of revenues among the top companies

# Section 4: Data Science

### Correlation Matrix

```python
numerical_cols = ["Rank", "Revenue", "Profits", "Employees"]

df_num = df[numerical_cols]

df_num.corr()
```

### We try to find the relationship between Employees(x) with Profits(y)

```python
df.plot.scatter(x="Employees",y="Profits")
```

#### Path1: Simple Regression

```python
slope = df_num.corr()["Employees"]["Profits"] * (df_num["Profits"].std())/(df_num["Employees"].std())
slope
```

```python
intercept = df_num["Profits"].mean()-slope*df_num["Employees"].mean()
intercept
```

```python
import seaborn as sns
sns.regplot(data=df, x="Employees", y="Profits", ci=None, line_kws={'color':'red'})
```

```python
df = df.drop(columns="Employees_K")
df["predict_profits"] = slope*df["Employees"] + intercept
df
```

```python
df['residual'] = df["Profits"] - df["predict_profits"]
df
```

#### Part2:Using Linear Regression in sklean

```python
# It will have the same predict value that we calculate manually before.
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df_num[["Employees"]],df["Profits"])
model.predict(df[["Employees"]])
df["predict2_profits"] = model.predict(df[["Employees"]])
df
```

#### Part3:Hypothesis Tests

- Null Hypothesis (H0): There is no linear correlation between the number of employees and profits.
- Alternative Hypothesis (H1): There is a linear correlation between the number of employees and profits.

We will use a significance level (α) of 0.05. If the p-value is less than α, we will reject the null hypothesis, suggesting that there is a significant linear relationship between the number of employees and profits.

```python

from scipy.stats import pearsonr
correlation, p_value = pearsonr(df['Employees'], df['Profits'])

print("Pearson Correlation:", correlation)
print("P-value:", p_value)

if p_value < 0.05:
    print("We reject the null hypothesis: There is a significant linear relationship between the number of employees and profits.")
else:
    print("We fail to reject the null hypothesis: There is no significant linear relationship between the number of employees and profits.")
```

### Thoughts about Section 4
- We explored the relationship between the number of employees and profit, using both linear regression and hypothesis test, we can conclude that they have some linear relationship. (Although their correlation coefficients are not very large).
